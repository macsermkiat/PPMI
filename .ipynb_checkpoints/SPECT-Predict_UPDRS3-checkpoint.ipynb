```python
# loop through all the DICOM files
image_array_2634 =[]
#for i in [38,39,40,41,42,43]:
for filenameDCM in lstFilesDCM:
    print(lstFilesDCM.index(filenameDCM))
    # read the file
    ds = pydicom.read_file(filenameDCM)
    # store the raw image data
    im2628 = np.vstack([ds.pixel_array[26],ds.pixel_array[27],ds.pixel_array[28]])
    im2931 = np.vstack([ds.pixel_array[29],ds.pixel_array[30],ds.pixel_array[31]])
    im3234 = np.vstack([ds.pixel_array[32],ds.pixel_array[33],ds.pixel_array[34]])
    im2634 = np.hstack([im2628,im2931,im3234])
    image_array_2634.append(im2634)
    #For debug corrupted image
    print(np.asarray(image_array_2634).shape)
```
